```python
import pandas as pd

# criteri per l'assegnazione dei tag
tag_criteria = {
    'Arte': {
        'churches': 2,
        'museums': 3,
        'castles': 2,
        'artworks': 5,
        'theatres': 2,
        'monuments': 3,
        'amphitheatres': 1,
    },
    'Cibo': {
        'restaurants': 10,
        'pubs': 1,
        'bars_cafes': 3,
        'biergarten': 3,
        'fast_foods': 4,
        'food_courts': 2,
        'ice_cream': 2
    },
    'Sport': {
        'gyms': 2,
        'swimming_pools': 2,
        'sports_halls': 2,
        'tennis_fields': 2,
        'padel_fields': 1,
        'soccer_fields': 3,
        'stadiums': 1,
        'sport_shops': 1,
    },
    'Natura': {
        'parks': 5,
        'picnic_areas': 2,
        'rivers': 1,
        'lakes': 1,
        'trails': 3
    },
    'Shopping': {
        'markets': 5,
        'supermarkets': 5,
        'clothes_shops': 3,
        'electronics_shops': 3,
        'jewelry_shops': 2
    },
    'Relax': {
        'parks': 3,
        'picnic_areas': 2,
        'spa': 1
    }
}

# Funzione per assegnare i tag in base ai criteri dettagliati
def assegna_tag(row):
    tags = []
    for tag, features in tag_criteria.items():
        total_features = sum(row[feature] >= criteria for feature, criteria in features.items())
        if total_features > 0:
            tags.append(tag)
    
    return ', '.join(tags)

test = True

if test:
    nome_file_input = 'Dataset/RISULTATO_CORTO.csv'
else:
    nome_file_input = 'Dataset/RISULTATO.csv'

df = pd.read_csv(nome_file_input)

# Applica la funzione agli elementi del DataFrame per assegnare i tag
df['Tags'] = df.apply(assegna_tag, axis=1)

# Ora il DataFrame contiene una colonna "Tags" che indica i tag assegnati a ciascun luogo

# Salva il DataFrame con i tag in un nuovo file o utilizzalo per ulteriori analisi
df.to_csv('Dataset/luoghi_con_tag.csv', index=False)
```

```python
# Definisci il nome del file di input e del file di output
file_di_input = "Dataset/luoghi_con_tag.csv"
file_di_output = "Dataset/DUMP_tags.sql"

# La stringa da aggiungere all'inizio e alla fine di ciascuna riga
stringa_iniziale = "INSERT INTO places VALUES( "
stringa_finale = ");"

i = 0
# Apre il file di input in modalità lettura e il file di output in modalità scrittura
with open(file_di_input, "r") as input_file, open(file_di_output, "w") as output_file:
    # Legge il file di input riga per riga
    for riga in input_file:
        # Concatena la stringa iniziale e finale alla riga
        riga_modificata = stringa_iniziale + riga.strip().replace("\"", "") + stringa_finale
        # Scrive la riga modificata nel file di output

        if i > 0:
            output_file.write(riga_modificata + "\n")
        
        i += 1

print(f"Il file '{file_di_output}' è stato creato con successo.")
```
