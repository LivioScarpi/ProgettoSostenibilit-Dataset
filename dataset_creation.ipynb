```python
import overpy

# Definisci la tua bounding box (latitudine e longitudine sud-ovest e nord-est)
bbox = (45.7692971, 7.815032400000001, 45.7892971, 7.8350324)

def get_informations_of_nature(bbox):
    api = overpy.Overpass()

    # Query Overpass API per ottenere informazioni su fiumi, laghi e sentieri naturali nella bounding box,
    # escludendo fontane, impianti idrici e sentieri asfaltati
    query = f"""
    [out:json];
    (
      way["natural"="water"]{bbox};
      way["waterway"="river"]{bbox};
      way["route"="hiking"]{bbox};
    );
    (
      ._;
      >;
    );
    out center;
    """

    try:
      time.sleep(2)
      result = api.query(query)

      # Inizializza contatori
      num_rivers = 0
      num_lakes = 0
      num_trails = 0

      # Elabora i risultati
      for way in result.ways:
          if way.tags.get('natural') == 'water':
              num_lakes += 1
          elif way.tags.get('waterway') == 'river':
              num_rivers += 1
          elif way.tags.get('route') == 'hiking':
              num_trails += 1

      return {"num_rivers": num_rivers, "num_lakes": num_lakes, "num_trails": num_trails}
    except Exception as e:
        return f"Errore: {str(e)}"

# Esegui la query
res = get_informations_of_nature(bbox)

print(res)

# Stampa i risultati
print(f"Numero di fiumi: {res['num_rivers']}")
print(f"Numero di laghi: {res['num_lakes']}")
print(f"Numero di sentieri naturali: {res['num_trails']}")
```

```python
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut

def do_geocode(address):
    geolocator = Nominatim(user_agent="myGeocoder", timeout=999999999)
    try:
        time.sleep(1)
        return geolocator.geocode(address)
    except GeocoderTimedOut:
        return do_geocode(address)

def get_bbox_from_location_name(location_name):
    # Inizializza il geolocalizzatore Nominatim
    #¬†geolocator = Nominatim(user_agent="myGeocoder", timeout=99999)

    # Ottieni le coordinate geografiche del luogo
    #location = geolocator.geocode(location_name + ", Italy")

    location = do_geocode(location_name + ", Italy")


    if location is not None:
        place_in_correct_region = False
        region = ""
        # Estrai le coordinate
        #print(location.raw.get('display_name').split(","))

        address = location.raw.get('display_name').split(",")

        for elem in address:
            elem_lower = elem.lower()
            if "piemonte" in elem_lower or "valle d'aosta" in elem_lower or "liguria" in elem_lower:
                #print(elem)
                #address = elem
                place_in_correct_region = True
                region = elem
                break

        lat = location.latitude
        lon = location.longitude

        # Calcola la BBOX
        delta = 0.01  # Imposta la dimensione della BBOX (puoi modificarla a tuo piacimento)
        bbox_inverted = (lon - delta, lat - delta, lon + delta, lat + delta)

        bbox = (bbox_inverted[1], bbox_inverted[0], bbox_inverted[3], bbox_inverted[2])
        coordinates = (lon, lat) # Inverti lat e lon per ottenere la posizione corretta

        return bbox, place_in_correct_region, region, coordinates
    else:
        print(f"Luogo '{location_name}' non trovato.")
        return None

# # Esempio di utilizzo:
# location_name = "Torino"
# bbox = get_bbox_from_location_name(location_name)

# if bbox:
#     print(f"BBOX per {location_name}: {bbox}")
```
